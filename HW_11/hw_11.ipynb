```python
#Даны две функции: 
#f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
#Требуется:
#1.Найти координаты точек пересечения
#2.Построить графики функций в одной системе координат
#3.Построить график функции пересечения
```

```python
from sympy import *
import matplotlib
```

```python
x = Symbol('x')
```

```python
f = x ** 3 - 50 * x
g = -x ** 4 + 88 * x ** 2 - 241
z = f - g
```

```python
roots = solve(z, x)
roots
```

```python
#1.Найти координаты точек пересечения
cross_points = {}
i = 1
for j in roots:
    cross_points[i] = {'x':round(N(j), 2), 'y': round(N(f.subs(x,j)), 2)}
    i += 1
cross_points
```

```python
#2.Построить графики функций в одной системе координат
plot(f, g)
```

```python
#3.Построить график функции пересечения
plot(z)
```
